```python
import pandas as pd
```

# .loc
Can be used to access the index of a row

```python
# Creating Series
animals = pd.Series(["Dog", "Cat", "Bird", "Panda", "Fish"], index=[1, 7, 7, 8, 3])
```

```python
animals
```

```python
# Accessing the index of Panda
animals.loc[7]  # Row Position
```

## .iloc
Can be used to access the position of a row

```python
animals = pd.Series(["Dog", "Cat", "Bird", "Panda", "Fish"])
```

```python
animals
```

```python
# Accessing the index of Fish
animals.iloc[4]  # Row Index
```
