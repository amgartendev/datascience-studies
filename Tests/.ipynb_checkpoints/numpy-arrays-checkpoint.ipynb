## Creating Numpy Arrays

```python
import numpy as np
```

## Array
Create an array

```python
np.array([1, 2, 3, 4, 5])
```

## Arange
Create an array with the given intervals

```python
np.arange(start=0, stop=15, step=3)
```

```python
# Inverting the first example
np.arange(start=15, stop=0, step=-3)
```

## Randint
Create an array with size N using pseudo-random numbers from low (lowest number INCLUSIVE) to high (highest number EXCLUSIVE)

```python
np.random.randint(low=0, high=10, size=5)
```

## Random
Create an array with size N using pseudo-random numbers between 0 and 1

```python
np.random.random(3)
```

## Rand
Create an array with the given shape

```python
np.random.rand(2, 2, 3)
```

## Zeros
Create an array in the given shape filled with zeros

```python
np.zeros((2, 5))
```

## Ones
Create an array in the given shape filled with ones

```python
np.ones((2, 5))
```
