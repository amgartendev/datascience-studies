```python
import pandas as pd
```

```python
car_sales = pd.read_csv('../car-sales.csv')
```

```python
car_sales
```

## Using .head() to see the top of the file
We can use the .head() function to see the top of our dataframe

```python
car_sales.head()
```

```python
# Specify how many rows we want to see
car_sales.head(7)
```

# Using .tail() to see the bottom of the file
We can use the .tail() function to see the tail of our dataframe

```python
car_sales.tail()
```

```python
# Specify how many rows we want to see
car_sales.tail(7)
```

# Using .sum()
We can use .sum() to see the sum of each column that contains integer values. String values will be concatenated

```python
car_sales.sum()
```

## Using .mean()
We can use .mean() to see the mean value of each column that contains integer values.

```python
# Deprecated Method
car_sales.mean()
```

```python
# Recommended Method
car_sales.mean(numeric_only=True)  # Specify numeric_only argument to calculate the mean of numeric values only
```

## Using .describe()

```python
car_sales.describe()
```
