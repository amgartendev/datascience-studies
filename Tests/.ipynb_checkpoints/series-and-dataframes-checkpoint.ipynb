## Creating Series

```python
# Importing libraries
import pandas as pd
```

```python
house_series = pd.Series(["Beedroom", "Living Room", "Bathroom"])
house_series
```

```python
colours_series = pd.Series(["Black", "Brown", "White"])
colours_series
```

## Creating DataFrames

```python
# Creating DataFrames
dataframe = pd.DataFrame("Rooms": house_series, "Colour": colours_series);
dataframe
```

## Correction of the cell above

Dataframe takes dictionaries as parameter, so we need to pass a dictionary as argument, that was what led the error in the cell above

```python
dataframe = pd.DataFrame({"Room": house_series, "Colour": colours_series})
dataframe
```
