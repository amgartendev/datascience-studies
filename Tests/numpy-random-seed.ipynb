# Numpy Random Seed
The function numpy.random.seeds(N) can be used to generate pseudo-random numbers that won't update when we run our cell. It can be used when we want to share our code with someone else and we want to reproduce the same experiment.

```python
import numpy as np
```

## Example of Random

```python
np.random.random(size=(5, 3))
```

We can see that every time we run our cell, the values within the array will change.

## Example of Random Seed

```python
np.random.seed(5)
np.random.random(size=(5, 3))
```

Now, if we run our cell multiple times, the values within the array will always be the same as long as the value parameter doesn't change
