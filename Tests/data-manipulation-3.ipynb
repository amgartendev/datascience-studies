```python
import pandas as pd
```

```python
car_sales = pd.read_csv('../car-sales.csv')
```

```python
car_sales
```

## Creating a New Column with Series

```python
total_fuel_used = pd.Series([4.7, 5.1, 3.6, 9.5, 4.5])
car_sales["Total Fuel Used"] = total_fuel_used
```

```python
car_sales
```

## Filling NaN with mean()

```python
car_sales["Total Fuel Used"] = car_sales["Total Fuel Used"].fillna(car_sales["Total Fuel Used"].mean())
```

```python
car_sales
```

## Creating a New Column with Python List

```python
fuel_capacity = [15.3, 11.2, 9.5, 16.0]
car_sales["Fuel Capacity"] = fuel_capacity
```

## Fixing Error
Different from pd.Series, when we are creating a new column using Python Lists, all rows have to be filled

```python
fuel_capacity = [15.3, 11.2, 9.5, 16.0, 11.0, 12.4, 16.6, 18.9, 14.4, 15.6]
car_sales["Fuel Capacity"] = fuel_capacity
car_sales
```

## Creating a new Column with a Single Value

```python
total_windows = 6
car_sales["Total Windows"] = total_windows
```
